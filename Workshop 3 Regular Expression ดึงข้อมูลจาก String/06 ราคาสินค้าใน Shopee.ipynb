## ราคาสินค้า

ประกอบไปด้วย ฿ และ , ที่ทำให้ไม่สามารถแปลงเป็นข้อมูลตัวเลขได้ ต้องทำการดึงข้อมูลส่วนตัวเลขออกมาและทำการแปลงเพื่อให้กลายเป็นตัวเลขใช้ประกอบการคำนวน

## ข้อมูลที่ใช้

นำเข้า library

```python
import re
import pandas as pd
```

ข้อมูลใน Excel

```python
table = pd.read_excel(r'C:\Users\acer\Desktop\Python Web Scraping\Selenium\ทีวี 55 นิ้ว 20 หน้าแรกบน shopee.xlsx')
table = table.set_index('title')
table
```

## pattern ราคา

สำหรับข้อมูลตัวเลขที่ต้องการคือเลขและ , ที่อยู่ติดกันโดยดึงมาได้แล้วค่อยลบ , ออกได้

#### pattern ของเลข 0-9 และ , ที่อยู่ติดกัน

```python
pattern = r'[0-9,]+'
pattern
```

```python
all_price_list = []
for price in table['price']:
    all_price = re.findall(pattern,price)
    price_list = []
    for p in all_price:
        price_int = int(p.replace(',',''))
        price_list.append(price_int)
    all_price_list.append(price_list)
```

## แปลงเป็น DataFrame

เพื่อเตรียมข้อมูลก่อนรวมเป็นตารางเดียวกันกับตารางเดิม การแปลงเป็น DataFrame นอกจากรวมข้อมูลง่ายยังเตรียมข้อมูลให้พร้อมเลย

#### สร้าง DataFrame() เก็บข้อมูลราคามาใส่ตัวแปร

```python
price_df = pd.DataFrame(all_price_list)
price_df.head(5)
```

#### ตั้ง index ให้ตรงกับข้อมูลที่จะไปรวมเดิม

```python
price_df.index = table.index
price_df.head(5)
```

#### ตั้ง column ให้มีชื่อที่เข้าใจจัดการง่าย

```python
price_df.columns = ['price min','price max']
price_df.head(5)
```

## นำตารางมาต่อกัน

#### ใช้ DataFrame มาต่อกันด้วยคำสั่ง .concat() ซึ่ง pandas จะจับคู่ row ที่ตรงกันให้

```python
report = pd.concat([table,price_df],axis=1)
report.head(5)
```

## บันทึกเป็นไฟล์ Excel

#### ใช้ DataFrame มาต่อกันด้วยคำสั่ง .concat() ซึ่ง pandas จะจับคู่ row ที่ตรงกันให้
