```python
import re
import pandas as pd
```

 Excel

```python
table = pd.read_excel(r'C:\Users\acer\Desktop\Python Web Scraping\Selenium\ 55  20  shopee.xlsx')
table = table.set_index('title')
table
```

## pattern 

 ,  , 

#### pattern  0-9  , 

```python
pattern = r'[0-9,]+'
pattern
```

```python
all_price_list = []
for price in table['price']:
    all_price = re.findall(pattern,price)
    price_list = []
    for p in all_price:
        price_int = int(p.replace(',',''))
        price_list.append(price_int)
    all_price_list.append(price_list)
```

##  DataFrame

  DataFrame 

####  DataFrame() 

```python
price_df = pd.DataFrame(all_price_list)
price_df.head(5)
```

####  index 

```python
price_df.index = table.index
price_df.head(5)
```

####  column 

```python
price_df.columns = ['price min','price max']
price_df.head(5)
```

## 

####  DataFrame  .concat()  pandas  row 

```python
report = pd.concat([table,price_df],axis=1)
report.head(5)
```
